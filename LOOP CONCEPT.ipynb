```python
#FOR LOOP      -----------------------A loop that repeats a block of code a fixed number of times.
for i in range(1, 11):
    print(i)
```

```python
#FOR LOOP ON LIST
fruits = ["apple", "banana", "mango"]

for f in fruits:
    print(f)
```

```python
#Print each mark.
marks = [45, 55, 60, 75, 90]

for m in marks:
    print(m)
```

```python
#2 WHILE LOOP
#Used when you don’t know how many times, but run until a condition becomes False.
count = 1
while count <= 5:
    print(count)
    count += 1

    
```

```python
i = 5

while i >= 1:
    print(i)
    i -= 1
```

```python
#Sum numbers until total reaches 20
total = 0

while total < 20:
    total += 5
    print("Total =", total)
```

```python
#What is enumerate()?

#enumerate() is used when you want:

#index (position)
#value (element)

#at the same time while looping.

#Normal for loop (WITHOUT enumerate)
fruits = ["apple", "banana", "mango"]

for f in fruits:
    print(f)
```

```python
#FOR loop WITH enumerate():
fruits = ["apple", "banana", "mango"]

for index, value in enumerate(fruits):
    print(index, value)
```

```python
#Print student marks with index
marks = [45, 55, 60, 75, 90]

for i, m in enumerate(marks):
    print("Index:", i, "Value:", m)
```

```python
#Start index from 1
marks = [45, 55, 60, 75, 90]

for i, m in enumerate(marks, start=1):
    print("Student", i, "scored", m)
```

```python
#NESTED LOOPS 

#A nested loop means a loop inside another loop.




matrix = [
    [1, 2, 3],
    [4, 5, 6]
]

for row in matrix:
    for num in row:
        print(num)
```

```python
#Print a Table (Rows & Columns)
for row in range(1, 4):        # outer loop → rows
    for col in range(1, 4):    # inner loop → columns
        print(row, col)
```
