```python
#Create a list and print it
fruits = ["apple", "banana", "mango"]
print(fruits)
```

```python
#Access the first and last element                                             #Index in List (Python)

                                                                               # An index is the position number of an item in a list.

                                                                                #    Python lists are zero-indexed, which means the first item has index 0, the second item has index 1, and so on.
fruits = ["apple", "banana", "mango"]
print(fruits[0])     
print(fruits[-1])    
```

```python
#Change one item in the list
colors = ["red", "blue", "green"]
colors[1] = "yellow"
print(colors)
```

```python
#Add a new element to the list                                              #list_name.append(value)

numbers = [1, 2, 3]
numbers.append(4)
print(numbers)
```

```python
#Insert an element at a specific position                                  #list_name.insert(index_position, value_to_insert)
animals = ["dog", "cat", "cow"]
animals.insert(1, "goat")
print(animals)
```

```python
#Remove an item from the list                                              #list_name.remove(value_to_remove)

items = ["pen", "pencil", "eraser"]
items.remove("pencil")
print(items)
```

```python
#Loop through a list
cities = ["Delhi", "Mumbai", "Chennai"]
for c in cities:
    print(c)
```

```python
#Find the length of a list
names = ["Divya", "Mohan", "Kiran"]
print(len(names))
```

```python
#Sort a list in ascending order
nums = [4, 2, 9, 1]                                           #sorted_list = sorted(list_name)
nums.sort()
print(nums)
```

```python
# descending 
marks = [55, 12, 98, 34]                                        #list_name.sort(reverse=True)
result = sorted(marks, reverse=True)
print(result)
```

```python
#Create a list of numbers and print only even numbers
numbers = [1, 2, 3, 4, 5, 6]

for n in numbers:
    if n % 2 == 0:
        print(n)
```

```python
#Join two tuples
t1 = (1, 2, 3)
t2 = (4, 5, 6)
combined = t1 + t2
print(combined)
```

```python
#Convert tuple into list, modify it, and convert back
t = (10, 20, 30)
temp = list(t)
temp.append(40)
t = tuple(temp)
print(t)
```

```python
#append() – Add one item at the end                                              #list_name.append(item)

fruits = ["apple", "banana"]
fruits.append("mango")
print(fruits)
```

```python
#extend() – Add multiple items
numbers = [1, 2, 3]
numbers.extend([4, 5, 6])
print(numbers)
```

```python
#insert() – Add item at a specific index                                 

colors = ["red", "blue", "green"]
colors.insert(1, "yellow")                    
print(colors)
```

```python
#remove() – Remove a specific item
items = ["pen", "pencil", "eraser"]
items.remove("pencil")
print(items)
```

```python
#pop() – Remove item using index                                         #list_name.pop(index_position)
students = ["Amit", "Raj", "Sara", "John"]
students.pop(2)
print(students)

```

```python
#Check in a list
fruits = ["apple", "banana", "mango"]
print("apple" in fruits)   
print("orange" in fruits)  
```
