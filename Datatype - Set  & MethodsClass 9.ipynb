```python
# basic operations-len()

set1={12,34,56,8,90}

print(len(set1))
```

```python
# built in methods in set
# Adding an element to set - add()

set1={12,34,56,8,90}
set1.add(10)
print(set1)
```

```python
set1={12,34,56,8,90}
print(set1)
```

```python
set1={12,34,56,8,90}
set1.add("python",10)
print(set1)
```

```python
# Adding multiple element to set - update()

set1={12,34,56,8,90}

set1.update(("python",100))
print(set1)
```

```python
set1={12,34,56,8,90}
set1.update(["hello","python"])
print(set1)
```

```python
set1={12,34,56,8,90}
set1.update({"hello","python"},[1,2,3])
print(set1)
```

```python
s1={12,32}
s2={34,56}

s1.update(s2)
print(s1)
```

```python
s1={12,32}
s2={34,56}
s2.update(s1)
print(s2)
```

```python
s1={12,32}
s2={34,56}
s3={"hello"}
s4={11,45}
S5={0,4,5,67}
s1.update(s2,s3,s4,S5)
print(s1)
```

```python
# removing element from set
# remove()

r1={"hello","python",12,34,56,889,9.8}
r1.remove(2)
print(r1)
```

```python
# discard()

d1={"hello","python",12,34,56,889,9.8}
d1.discard(2)
print(d1)
```

```python
# pop()

p1={"hello","python",12,34,56,889,9.8}

p1.pop()
print(p1)
```

```python
# union()

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
a.union(b)
```

```python
a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
c={"red",'black','pink'}
a.union(b,c)
```

```python
# intersection()

a={'yellow','black','pink'}
b={'pink','white','grey','green','black'}

a.intersection(b)
```

```python
a={'orange','yellow','black','pink',"green"}
b={'pink','white','grey','green','black'}
c={"green"}
a.intersection(b,c)
```

```python
# intersection_update()

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
print(a)
a.intersection_update(b)
print(a)
```

```python
a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
print(b)
b.intersection_update(a)
print(b)
```

```python
# difference()

a={'orange','yellow','black','pink',"blue"}
b={'pink','black','blue'}
a.difference(b)
```

```python
a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
b.difference(a)
```

```python
# difference_update()

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
print(a)
a.difference_update(b)
print(a)
```

```python
a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
b.difference_update(a)

print(b)
```

```python
# symmetric difference()

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
a.symmetric_difference(b)
```

```python
# symmetric difference_update()

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
a.symmetric_difference_update(b)
print(a)
```

```python
# issubset() a(b) 

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}

a.issubset(b)
```

```python
a={'black','pink',"yellow"}
b={'pink'}

b.issubset(a)
```

```python
# issuperset() a sup (b)

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
a.issuperset(b)
```

```python
a={'orange','yellow','pink',"white","green","grey"}
b={'pink','white','grey','green'}
a.issuperset(b)
```

```python
# isdisjoint() disjoint means not common elements

a={'orange','yellow','black','pink'}
b={'pink','white','grey','green','black'}
b.isdisjoint(a)
```

```python
a={'orange','yellow',"red"}
b={'pink','white','grey','green','black'}
a.isdisjoint(b)
```
