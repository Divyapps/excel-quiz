```python
# copy()

d1={"name":"sakthi","age":67,"height":4.5}

d2=d1.copy()

print(d2)
```

```python
# fromkeys

key=["sarathi","bharath","kannan"]

value=0

dict1={}

result=dict1.fromkeys(key,value)
print(result)
```

```python
# setdefault()

d={"name":"sakthi","age":67,"height":4.5}

d.setdefault("fname","shiva")
print(d)
```
