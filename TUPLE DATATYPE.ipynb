```python
#Create a tuple and print it
colors = ("red", "blue", "green")
print(colors)
```

```python
#Access elements using index
fruits = ("apple", "banana", "mango")
print(fruits[0])     # apple
print(fruits[2])     # mango
```

```python
#Check the length of a tuple
items = ("pen", "pencil", "eraser")
print(len(items))
```

```python
#Slice a tuple (print part of it)
nums = (10, 20, 30, 40, 50)
print(nums[1:4])     
```

```python
#Check if an item exists in a tuple
days = ("Mon", "Tue", "Wed")
print("Tue" in days)     
```

```python
#Count occurrences in a tuple
data = (1, 2, 2, 3, 4, 2)
print(data.count(2))     
```

```python
#Find the index of an item
cities = ("Delhi", "Mumbai", "Chennai")
print(cities.index("Mumbai"))
```

```python
#Loop through a tuple
animals = ("dog", "cat", "cow")
for a in animals:
    print(a)
```

```python
#Check in a tuple
numbers = (10, 20, 30)
print(20 in numbers)  
print(50 in numbers)  
```

```python
#len(tuple) – Number of elements
t = (10, 20, 30, 40)
print(len(t))


```

```python
#max(tuple) – Largest element
t = (5, 10, 2, 8)
print(max(t))
```

```python
#min(tuple)	-------------------Smallest element
#sum(tuple)	------------------Sum of all elements
print(min(t))
t = (1, 2, 3, 4)
print(sum(t))
```

```python
#tuple(iterable) – Convert iterable to tuple
lst = [1, 2, 3]
t = tuple(lst)
print(t)
```

```python
#tuple.index(value) – Index of first occurrence
t = (10, 20, 30, 20)
print(t.index(20))
```

```python
#tuple.count(value) – Count of a value
t = (1, 2, 2, 3, 2)
print(t.count(2))
```
