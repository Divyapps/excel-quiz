```python
#Add two float numbers
a = 10.5
b = 4.3
print(a + b)
```

```python
#Multiply two float numbers
x = 2.5
y = 3.2
print(x * y)
```

```python
#Take float input from user and print its square
n = float(input("Enter a decimal number: "))
print(n * n)
```

```python
#Calculate the area of a circle using float
radius = 4.5
area = 3.14 * radius * radius
print(area)
```

```python
#Add integer + float and print the result
a = 10      # int
b = 5.75    # float
print(a + b)
```

```python
#Find the average of three float numbers
x = 10.5
y = 20.2
z = 5.3

avg = (x + y + z) / 3
print(avg)
```

```python
#Convert hours to minutes using float
hours = 1.5
minutes = hours * 60
print(minutes)
```

```python
#Calculate simple interest (float)
p = 1500.50
r = 5.5
t = 2.0

si = (p * r * t) / 100
print(si)
```

```python

#round() – Round a float
x = 10.67
print(round(x))
```

```python
#type() – Check datatype
x = 10.5
print(type(x))
```

```python
#abs() – Absolute value                      #What abs() does?

                                           # abs() is a built-in Python function

                                            #   It returns the absolute value of a number

                                                 #   Absolute value means the distance from zero, ignoring the sign.
x = -12.55
print(abs(x))
```

```python
#max() – Largest float
print(max(2.5, 8.9, 3.3))
```

```python
#math.floor() – Round down
#math.ceil() – Round up
#import math
print(math.floor(4.9))   # 4
print(math.ceil(4.1))    # 5
```
