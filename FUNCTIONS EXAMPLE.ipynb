```python
# Function with Task;

def fun(lst):
    x=0
    y=0
    for i in lst:
        if i%2==0:
            x+=1
        else :
            y+=1
    else:
        return x,y


lst=[5,10,15,20,25,30,40] 

even,odd=fun(lst)
print("The even number value:",even)
print("The odd number value:",odd)
```
