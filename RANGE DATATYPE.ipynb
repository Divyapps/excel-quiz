```python
#RANGE() FUNCTION IN PYTHON

#range() is used to generate a sequence of numbers.
#It is commonly used in loops, lists, and number generation.
```

```python


#There are 3 ways to use range():

   #range(stop)

#Starts from 0 → goes up to stop - 1

    #range(start, stop)

#Starts from start → goes up to stop - 1
 ️ 
 #range(start, stop, step)

#Starts from start, increments by step
```

```python
#range(stop)
for i in range(5):
    print(i)
```

```python
#range(start, stop, step)                #Step = jump value
for i in range(1, 10, 2):
    print(i)
```

```python
#Negative Step
for i in range(10, 0, -1):
    print(i)
```
