```python
#len() – Number of items
student = {"name": "Divya", "age": 25, "course": "Python"}
print(len(student))
```

```python
#keys() – Get all keys
student = {"name": "Divya", "age": 25, "course": "Python"}
print(student.keys())
```

```python
#values() – Get all values
print(student.values())
```

```python
#items() – Get key-value pairs
print(student.items())
```

```python
#update() – Add or update items
student.update({"age": 26, "city": "Delhi"})
print(student)
```

```python
#pop() – Remove a key and return its value
age = student.pop("age")
print(age)
print(student)
```

```python
#popitem() – Remove last inserted item
student.popitem()                                              #popitem()
print(student)

                                                                #   Removes the last inserted key–value pair from the dictionary.

                                                                  # Returns the removed item (but here we didn’t store it).

                                                                    #     Dictionary becomes smaller by one item.

                                                                                  #    print(student)

                                                                                      #   Displays the dictionary after removing the last item.
```

```python
#clear() – Remove all items
student.clear()
print(student)
```
