```python
#Print your name using a string
name = "ANU"
print(name)
```

```python
#Concatenate (join) two strings
first = "Hello"
second = "Python"
print(first + " " + second)
```

```python
#Find the length of a string
text = "Learning Python"
print(len(text))
```

```python
#Convert a string to uppercase
word = "python"
print(word.upper())
```

```python
#Convert a string to lowercase
name = "COURSE"
print(name.lower())
```

```python
#Take a string input and print the first character
s = input("Enter a word: ")
print(s[0])
```

```python
#Slice a string (print substring)
text = "HelloPython"
print(text[0:5])   # Prints 'Hello'
```

```python
#Count occurrences of a letter in a string
s = "banana"                                                     #count() – Count occurrences
print(s.count("a"))
```

```python
#Replace a word in a string
sentence = "I like Java"
print(sentence.replace("Java", "Python"))
```

```python
#Check if a string starts with a given letter
msg = "Python Programming"
print(msg.startswith("p"))
```

```python
#title() – Converts to Title Case
text = "welcome to python"
print(text.title())
```

```python
#capitalize() – First letter capital                        #Converts the first letter of the string to uppercase
                                                          #Converts all other letters to lowercase
text = "python programming"
print(text.capitalize())
```

```python
#strip() – Remove spaces                                         
word = "   python   "                                                   
print(word.lstrip())                                           #lstrip() – Removes spaces from LEFT side only
word = "   python   "
print(word.rstrip())                                           #rstrip() – Removes spaces from RIGHT side only
word = "   python   "
print(word.strip())


```

```python
#split() – Convert string → list
data = "apple,banana,grapes"
print(data.split(","))
```

```python
#join() – Convert list → string
items = ['a', 'b', 'c']
print("-".join(items))
```

```python
#isalpha() – Check alphabet only
print("abc".isalpha())
```

```python
#isdigit() – Check digits only
print("12375".isdigit())
```

```python
#isalnum() – Alphabets + numbers
print("abc123".isalnum())
```
