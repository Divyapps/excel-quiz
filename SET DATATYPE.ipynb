```python
#Create a set and print it
fruits = {"apple", "banana", "mango"}
print(fruits)
```

```python
#Add an element to a set
colors = {"red", "blue"}
colors.add("green")
print(colors)
```

```python
#Add multiple items using update()
nums = {1, 2, 3}
nums.update([4, 5])
print(nums)
```

```python
#Remove an item from a set
items = {"pen", "pencil", "eraser"}
items.remove("pencil")
print(items)
```

```python
items = {"pen", "pencil", "eraser"}
items.remove("pencil")
items.discard("marker")   
print(items)
```

```python
#Check if an item exists in the set
cars = {"BMW", "Audi", "Ford"}
print("Audi" in cars)     # True
```

```python
#Set union (combine two seTS)
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))
```

```python
#Set intersection (common items)
x = {10, 20, 30}
y = {20, 30, 40}
print(x.intersection(y))
```

```python
#Set difference
a = {1, 2, 3, 4}
b = {3, 4}
print(a.difference(b))   
```

```python
#Loop through a set
animals = {"dog", "cat", "cow"}
for a in animals:
    print(a)
```

```python
#add() – Add an element to the set
colors = {"red", "blue"}
colors.add("green")
print(colors)
```

```python
#clear() – Remove all elements
nums = {1, 2, 3}
nums.clear()
print(nums)
```

```python
#union() – Combine two sets
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))
```

```python
#difference() – Elements in first set but not in second
a = {1, 2, 3}
b = {2, 3, 4}
print(a.difference(b))
```

```python
#intersection() – Common elements
a = {1, 2, 3}
b = {2, 3, 4}
print(a.intersection(b))
```
