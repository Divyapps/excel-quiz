```python
#Function that prints Hello
def say_hello():
    print("Hello!")

say_hello()
```

```python
def greet():
    print("Hello, welcome to Python!")

greet()
```

```python
#Function that adds 2 numbers
def add(a, b):
    print(a + b)

add(10, 20)
```

```python
def add(a, b):
    result = a + b
    return result

print(add(5, 3))
```

```python
#Function that returns the square of a number
def square(num):
    return num * num

result = square(5)
print(result)
```

```python
#Function with default value
def greet(name="Student"):
    print("Hello", name)

greet()          # uses default
greet("Anu")   # custom name
```

```python
#Function that finds total marks
def total_marks(marks):
    return sum(marks)

marks = [45, 55, 60]
print(total_marks(marks))
```

```python
#Function that calculates average of a list
def average(values):
    return sum(values) / len(values)

scores = [10, 20, 30, 40]
print(average(scores))
```

```python
#Function that checks if number is even
def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False

print(is_even(10))
print(is_even(7))
```
