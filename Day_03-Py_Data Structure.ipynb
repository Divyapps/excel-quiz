### Python allows Swapping of Variables with out temp variable

```python
# Python program to swap two variables
# Using a temporary variable

x=5
y=10 


# To take inputs from the user
#x = input('Enter value of x: ')
#y = input('Enter value of y: ')
print('The value of x before swapping: {}'.format(x))
print('The value of y before swapping: {}'.format(y))
# create a temporary variable and swap the values

z=x #5
x=y #10
y=z #5

print(x)
print(y)

print('The value of x after swapping: {}'.format(x))
print('The value of y after swapping: {}'.format(y))
# Python program to swap two variables
# With out Using a temporary variable
```

```python
x = 5
y = 10
print('The value of x before swapping: {}'.format(x))
print('The value of y before swapping: {}'.format(y))
# The following code swapping but without the use of any temporary variable.
x = y
print("x =", x)
print("y =", y)
print('The value of x after swapping: {}'.format(x))
print('The value of y after swapping: {}'.format(y))
```

```python
# Python3 code to demonstrate
# to convert list of string and characters
# to list of characters
# using list comprehension

# initializing list
test_list = ['abcd', 'i', 's', 'be', 's', 't']

# printing original list
print("The original list is : " + str(test_list))

# using list comprehension
# to convert list of string and characters
# to list of characters
res = [i for ele in test_list for i in ele]

# printing result
print("The list after conversion is : " + str(res))
```
