```python
# what is list? list is mutable
# what is tuple? tuple is immutable
# mutable and immutable
```

```python
# mutable list (adding an element)


l2=[10,20,30,40]

l2[0]="hai"

print(l2)
```

```python
# immutable tuple (adding an element)

t2=(10,20,30,40)

t2[0]="hai"
print(t2)
```

```python
# mutable list (deleting an element)

l3=[10,20,30]

del l3[0]
print(l3)
```

```python
# immutable tuple (deleting an element)

t3=(10,20,30)

del t3[0]
print(t3)
```

```python
# packing and unpacking 

# Unpacking
a=("sakthi",23,"male")

name,age,gender=a

print(name)
print(age)
print(gender)
```
